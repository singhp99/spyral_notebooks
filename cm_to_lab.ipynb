```python
import numpy as np
import vector
from spyral_utils.nuclear import NuclearDataMap

ejectile_z = 1
ejectile_a = 2
projectile_z = 6
projectile_a = 15
target_z = 1
target_a = 2

residual_z = target_z + projectile_z - ejectile_z
residual_a = target_a + projectile_a - ejectile_a

nuclear_map = NuclearDataMap()

ejectile = nuclear_map.get_data(ejectile_z, ejectile_a)
projectile = nuclear_map.get_data(projectile_z, projectile_a)
target = nuclear_map.get_data(target_z, target_a)
residual = nuclear_map.get_data(residual_z, residual_a)

proj_energy_start = 106.5
target_vector = vector.array({"px": [0.0], "py": [0.0], "pz": [0.0], "E": [target.mass]})

data = np.loadtxt("/Users/pranjalsingh/Desktop/inel_data.txt") #change the data file here
cm_angles_deg = data[:, 0]   
cm_angles = np.deg2rad(cm_angles_deg)

beam_vec_max = vector.array({
    "px": [0.0],
    "py": [0.0],
    "pz": [np.sqrt(proj_energy_start * (proj_energy_start + 2.0 * projectile.mass))],
    "E": [proj_energy_start + projectile.mass]
})

parent_max = target_vector + beam_vec_max
parent_cm_max = parent_max.boostCM_of(parent_max)
ex = 0.0 #enter the excitation state 
eject_e_cm = (ejectile.mass**2 - (residual.mass + ex)**2 + parent_cm_max.E**2) / (2.0 * parent_cm_max.E)
eject_p_cm = np.sqrt(eject_e_cm**2 - ejectile.mass**2)

eject_vec_cms = vector.array({
    "px": eject_p_cm * np.cos(np.pi-cm_angles),
    "py": np.zeros_like(np.pi-cm_angles),
    "pz": eject_p_cm * np.cos(np.pi-cm_angles),
    "E": np.full_like(np.pi-cm_angles, eject_e_cm)
})

eject_vec_lab = eject_vec_cms.boost(parent_max)
lab_angles_deg = np.rad2deg(eject_vec_lab.theta)

print(lab_angles_deg)
```
